Import necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load dataset in csv format

```python
spotifydf=pd.read_csv("C:\\Users\\SHUBHAM KADAM\\SHUBHAM PROJECTS\\Data analytics projects\\Spotify\\spotify dataset.csv")
#information about spotifydf
spotifydf.info()

#intial 5 values of spotifydf
spotifydf.head()
```

Descriptive analysis of spotifydf dataframe

```python
#Descriptive analysis of spotidf
print(f"DESCRIPTIVE ANALYSIS OF spotidf \n{spotifydf.describe()}")
```

Removing duplicates

```python
#Removing duplicates from spotidf
spotifydf=spotifydf.dropna().drop_duplicates()
print(spotifydf.info())
```

Descriptive analysis of spotifydf dataframe after deduplication

```python
#Descriptive analysis of newspotidf
print(spotifydf.describe())
```

Most interacted users

```python
#Getting top 5 users by the MLsec_played
print("\nMOST ACTIVE USERS")
spotifydf.nlargest(5,'MLsec_played')
```

Some common analysis

```python
#Total number of artists
artist=spotifydf.groupby('Artist_name')['User_no'].count()
print(f"TOTAL NO OF ARTISTS ARE {artist.count()} \n")

#Total number of albums
album=spotifydf.groupby('Album_name')['User_no'].count()
print(f"TOTAL NO OF ALBUMS ARE {album.count()} \n")

#Reason to start 
rsnstart=spotifydf.groupby('Reason_start')['User_no'].count()
print(f'REASON TO START LISTENING\n {rsnstart} \n')

#Reason to end
rsnend=spotifydf.groupby('Reason_end')['User_no'].count()
print(f'REASON TO END LISTENING\n {rsnend}')
```

Top 10 Albums

```python
albmsdf=spotifydf.groupby('Album_name')['User_no'].count().sort_values(ascending=False).head(10)
print(f"Top 10 Albums\n {albmsdf}")
```

Top 10 Artists

```python
artistdf=spotifydf.groupby('Artist_name')['User_no'].count().sort_values(ascending=False).head(10)
print(f"Top 10 Artists\n {artistdf}")
```

Users by Platform

```python
usernew={'Android':spotifydf[spotifydf['Platform']=='android'].shape[0],'Cast to device':spotifydf[spotifydf['Platform']=='cast to device'].shape[0],'Other':spotifydf[spotifydf['Platform'].isin(['iOS','web player','mac','windows'])].shape[0]}
usernew=pd.Series(usernew)
print(usernew)
plt.Figure(figsize=(10,5))
plt.title("Platform uses Distribution")
plt.pie(usernew,labels=usernew.index,autopct='%1.1f%%')
plt.legend()
plt.show()
```

Shuffled users

```python
shuffledf=spotifydf['Shuffle'].value_counts()
plt.Figure(figsize=(10,5))
shuffledf.plot(kind='bar',color=['red','blue'])
plt.title("User shuffled or Not")
plt.ylabel("Number of users")
plt.xticks(rotation=0)
plt.show()
```

Reason behind starting Spotify

```python
rsnstartdf=spotifydf.groupby('Reason_start')['User_no'].count().reset_index()
plt.Figure(figsize=(10,5))
rsnstartdf['Reason_code']=rsnstartdf['Reason_start'].astype('category').cat.codes
sns.scatterplot(data=rsnstartdf,x='Reason_code',y='User_no')
plt.xticks(ticks=rsnstartdf['Reason_code'],labels=rsnstartdf['Reason_start'],rotation=45) 
plt.xlabel("Reason for starting")
plt.title("Why users starts using spotify")
plt.grid(True)
plt.show()
```

Why stop listening

```python
plt.Figure(figsize=(10,5))
rsnenddf=spotifydf.groupby("Reason_end")['User_no'].count().reset_index()
sns.lineplot(rsnenddf,x='Reason_end',y='User_no',color='red',markers='o')
plt.xlabel("Reason to stop listening")
plt.ylabel("User no")
plt.title("Why they stop listening")
plt.xticks(rotation=90)
plt.show()
```

Songs skipped or not

```python
skippeddf=spotifydf['Skipped'].value_counts()
skippeddf.plot(kind='bar',color=['orange','pink'])
plt.Figure(figsize=(10,5))
plt.title("User skipped or not")
plt.xlabel("Skipped")
plt.ylabel("Users")
plt.xticks(rotation=0)
plt.show()
```

Sum of milisec played by Platform

```python
mlsecplddf=spotifydf.groupby('Platform')['MLsec_played'].sum().reset_index()
print(mlsecplddf)
plt.Figure(figsize=(10,5))
mlsecplddf['reset1']=mlsecplddf['Platform'].astype('category').cat.codes
sns.scatterplot(data=mlsecplddf,y='MLsec_played',x='reset1',hue='MLsec_played',palette=['red','darkgreen','blue','black','darkorange','darkblue'])
plt.title("MLsec_played vs platform")
plt.xticks(ticks=mlsecplddf['reset1'],labels=mlsecplddf['Platform'])
plt.ylabel('MLsec played')
plt.grid(True)
plt.xlabel('Platform')
plt.show()
```
